```python
from FlagEmbedding import BGEM3FlagModel
model = BGEM3FlagModel('BAAI/bge-m3',  use_fp16=False) # Setting use_fp16 to True speeds up computation with a slight performance degradation

def embed_text(texts):
  output = model.encode(texts, return_dense=True, return_sparse=True, return_colbert_vecs=False)

  return output['dense_vecs'] , output

def prepare_df(records):
  from tqdm import tqdm
  for index, row in tqdm(records.iterrows()):
    dense, output = embed_text(row['ภาษากลาง'])
    records.at[index, 'dense_vecs'] = dense
    records.at[index, 'sparse_index'] = [int(key) for key in output['lexical_weights'].keys()]
    records.at[index, 'sparse_value'] = list(output['lexical_weights'].values())

  return records
```
